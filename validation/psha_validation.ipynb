# Validation of probabilistic seismic hazard analysis using ground motion models models

The purpose of this notebook is to validate our implementation of the probabilistic seismic hazard analysis against OpenSHA (Field, 2003). We have gathered screenshots of relevant inputs into OpenSHA, and are plotting the outputs from the GUI against our ucla_plha code. The OpenSHA code provides a probability of exceedance. To accurately compare the probability of exceedance to an annual rate of exceedance from our PSHA code we use the following Poisson distribution probability relationship. 

$P[IM > im] = 1-e^{-\lambda *t}$

Alignment of the image and the data plot is performed by visual inspection and iteration, and is not perfect.

Field, E.H., T.H. Jordan, and C.A. Cornell (2003), OpenSHA: A Developing Community-Modeling Environment for Seismic Hazard Analysis, Seismological Research Letters, 74, no. 4, p. 406-419.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()


ax.plot(pga_bssa, -np.log(1-exceedance_bssa),color="tab:blue", label="OpenSHA_bssa")



ax.plot(config_bssa['output']['psha']['PGA'],
        config_bssa['output']['psha']['annual_rate_of_exceedance'],
        label='ucla_plha BSSA14', color= "red")


ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-6, 1e0)


ax.set_xlabel('PGA [g]')
ax.set_ylabel(r'$\lambda (IM > im)$ $[yr^{-1}]$')
ax.set_title("Probabilistic Seismic Hazard Curve")


ax.grid(True, alpha=0.5, which='both')
ax.set_facecolor('xkcd:white')
ax.legend()


fig.set_figwidth(7)
fig.set_figheight(5)
plt.tight_layout()
plt.show()
```

### Note regarding Hazard Difference 

#### Background seismicity is gridded and fault ruptures are not explicitly modeled in UCERF3. We interpret the grid locations as being epicenters, and compute other distance metrics from epicentral distance using curve fits to the default methods described by Thompson and Worden (2017)

#### The CY2014 and ASK14 ground motion models show some deviations from the OpenSHA hazard curve. The difference in hazard curve can be attributed to fault distance calculations for the background seismicity. When calculating only the known fault plane solution, there is very good agreement between the hazard curves.
