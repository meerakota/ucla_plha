## Example notebook

#### This Jupyter notebook serves as an example of how to use the ucla_plha code.The ucla_plha code uses a Javascript Object Notation (JSON) input file to configure the analysis, and uses JSON Schema to validate user inputs. The example will be run for the 10 example config files we have created. Information about how to create and structure a config file is available in the README.md.Example config files are provided in the examples directory, and the schema is defined by ucla_plha_schema.json located in the src/ucla_plha directory.

#### This is a summary of the 10 example config files. 

| config file | description |
| -------- | ----------- |
| [config1.json](config1.json) | PSHA (no PLHA) using BSSA14  |
| [config2.json](config2.json) | PSHA (no PLHA) using ASK14 |
| [config3.json](config3.json) | PSHA (no PLHA) using CB14 |
| [config4.json](config4.json) | PSHA (no PLHA) using CY14 |
| [config5.json](config5.json) | PSHA (no PLHA) using BSSA14, ASK14, CB14, and CY14 |
| [config6.json](config6.json) | PSHA (no PLHA) including PSHA disaggregation using BSSA14, ASK14, CB14, and CY14 |
| [config7.json](config7.json) | PSHA and PLHA using BSSA14,ASK14,CB14, and CY14 and Boulanger and Idriss 2016 liquefaction triggering model|
| [config8.json](config8.json) | PSHA and PLHA using BSSA14,ASK14,CB14, and CY14 and Moss et al. 2006 liquefaction triggering model|
| [config9.json](config9.json) | PSHA and PLHA using BSSA14,ASK14,CB14, and CY14 and NGL SMT liquefaction triggering model|
| [config10.json](config10.json) | PSHA and PLHA including PSHA/PLHA disaggregation using BSSA14, ASK14, CB14, and CY14 and Boulanger and Idriss 2016 liquefaction triggering model, NGL SMT liquefaction triggering model, and Moss et al. 2006 liquefaction triggering model (also shows example for saving output file)|

#### Config 6- PSHA (no PLHA) including PSHA disaggregation using BSSA14, ASK14, CB14, and CY14

```python
config6= plha.get_hazard('config6.json')
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(config6['output']['psha']['PGA'],
        config6['output']['psha']['annual_rate_of_exceedance'],
        label='Average of the 4 GMMs')


ax.set_xscale('log'); ax.set_yscale('log')
ax.set_xlabel('pga [g]')
ax.set_ylabel(r'$\lambda (IM > im)\;[yr^{-1}]$')
ax.set_title('Probabilistic Seismic Hazard Curve')
ax.set_ylim(1e-5, 1e1)
ax.grid(True, which='both', alpha=0.3)
ax.set_facecolor('xkcd:white')
ax.legend(title='GMM', loc='lower left', frameon=True)

plt.tight_layout()

plt.show()
```

```python
import numpy as np

magnitude_bin_edges=np.asarray(config6['input']['output']['psha']['disaggregation']['magnitude_bin_edges'])
pga = np.asarray(config6['input']['output']['psha']['pga'])
distance_bin_edges=np.asarray(config6['input']['output']['psha']['disaggregation']['distance_bin_edges'])
epsilon_bin_edges=np.asarray(config6['input']['output']['psha']['disaggregation']['epsilon_bin_edges'])
m_center = 0.5 *( magnitude_bin_edges[0:-1] + magnitude_bin_edges[1:])
r_center = 0.5 * (distance_bin_edges[0:-1] + distance_bin_edges[1:])
epsilon_center = 0.5 * (epsilon_bin_edges[0:-1] + epsilon_bin_edges[1:])

pga_index = 8
_xx, _yy = np.meshgrid(r_center, m_center)
x, y = _xx.ravel(), _yy.ravel()
disagg = np.asarray(config6['output']['psha']['disaggregation'], dtype=float)

if(len(m_center)>1):
    depth = (m_center[1] - m_center[0]) * 0.15
else:
    depth = 1

if(len(r_center)>1):
    width = (r_center[1] - r_center[0]) * 0.15
else:
    width = 1
bottom = np.zeros(len(x))

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))

my_cmap = plt.get_cmap("RdYlBu")
def rescale(e):
    return (e - epsilon_bin_edges[1]) / (epsilon_bin_edges[-2] - epsilon_bin_edges[1])
alpha = 0.1
for i, e in enumerate(epsilon_center):
    top = disagg[pga_index,:, :, i].ravel()
    if(i==0):
        label = '<' + str(epsilon_bin_edges[1])
    elif(i==len(epsilon_center)-1):
        label = '>' + str(epsilon_bin_edges[-2])
    else:
        label = str(e-0.5) + ' to ' + str(e+0.5)
    ax.bar3d(x, y, bottom, width, depth, top, shade=True, label=label, zsort='max', edgecolor='black', facecolor=my_cmap(rescale(e)))
    bottom += disagg[pga_index,:, :, i].ravel()
ax.set_title('PGA = ' + str(np.round(pga[pga_index],3)) + 'g')
ax.legend(title = r'$\varepsilon$')
ax.set_xlabel('Distance [km]')
ax.set_ylabel('Magnitude')
ax.set_zlabel('% Contribution')
plt.tight_layout()
```

#### Config 10- PSHA and PLHA including PSHA/PLHA disaggregation using BSSA14, ASK14, CB14, and CY14 and Boulanger and Idriss 2016 liquefaction triggering model, NGL SMT liquefaction triggering model, and Moss et al. 2006 liquefaction triggering model (also shows example for saving output file)

```python
config10= plha.get_hazard('config10.json')
```

```python
import matplotlib.pyplot as plt
import numpy as np 
fig, axs = plt.subplots(1, 2, sharey=True)


axs[0].plot(config10['output']['psha']['PGA'],config10['output']['psha']['annual_rate_of_exceedance'], label = 'BSSA14, ASK14, CB14, and CY14 Average')


axs[0].legend(title='', loc='upper right', frameon=True)
axs[0].set_yscale('log')
axs[0].set_xscale('log')
axs[1].set_yscale('log')
axs[0].set_ylim(10e-6,10e0)
axs[1].plot(config10['output']['plha']['FSL'], config10['output']['plha']['annual_rate_of_nonexceedance'], label = 'NGL SMT Model 2024, B&I 2016, and Moss 2006')


axs[1].legend(title='', loc='lower right', frameon=True)

axs[0].set_xlabel('pga [g]')

axs[0].set_ylabel('$\lambda (IM > im)$ $[yr^{-1}]$')
axs[0].set_title("a) Probabilistic Seismic Hazard Curve")
axs[0].grid(True, alpha=0.3, which='both')
axs[0].set_facecolor('xkcd:white')

axs[1].set_title("b) Probabilistic Liquefaction Hazard Curve")

axs[1].set_ylabel('$\lambda (FS_L < fs_l) [yr^{-1}]$')
axs[1].set_xlim(0, 2)
axs[1].set_xlabel('$FS_{L}$ [-]')
axs[1].grid(True, alpha=0.3, which='both')



fig.set_figwidth(10)
fig.set_figheight(4)
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

magnitude_bin_edges = np.asarray(config10['input']['output']['psha']['disaggregation']['magnitude_bin_edges'])
distance_bin_edges  = np.asarray(config10['input']['output']['psha']['disaggregation']['distance_bin_edges'])
epsilon_bin_edges   = np.asarray(config10['input']['output']['psha']['disaggregation']['epsilon_bin_edges'])
pga = np.asarray(config10['input']['output']['psha']['pga'])
fsl = np.asarray(config10['input']['output']['plha']['fsl'])


m_center = 0.5 * (magnitude_bin_edges[:-1] + magnitude_bin_edges[1:])
r_center = 0.5 * (distance_bin_edges[:-1] + distance_bin_edges[1:])
epsilon_center = 0.5 * (epsilon_bin_edges[:-1] + epsilon_bin_edges[1:])

pga_index = 9
fsl_index = 9


depth = (m_center[1] - m_center[0]) * 0.2 if len(m_center) > 1 else 1.0
width = (r_center[1] - r_center[0]) * 0.2 if len(r_center) > 1 else 1.0

_xx, _yy = np.meshgrid(r_center, m_center)
x, y = _xx.ravel(), _yy.ravel()


disagg_psha = np.asarray(config10['output']['psha']['disaggregation'], dtype=float)
disagg_plha = np.asarray(config10['output']['plha']['disaggregation'], dtype=float)


my_cmap = plt.get_cmap("RdYlBu")

def rescale_for_cmap(e):
    hi = epsilon_bin_edges[1] if len(epsilon_bin_edges) > 1 else epsilon_bin_edges[0]
    lo = epsilon_bin_edges[-2] if len(epsilon_bin_edges) > 2 else epsilon_bin_edges[-1]
    return 0.5 if hi == lo else (e - lo) / (hi - lo)

def epsilon_label(i, e):
    if i == 0 and len(epsilon_bin_edges) > 1:
        return f"<{epsilon_bin_edges[1]}"
    elif i == len(epsilon_center) - 1 and len(epsilon_bin_edges) > 2:
        return f">{epsilon_bin_edges[-2]}"
    else:
        return f"{e-0.5} to {e+0.5}"


fig = plt.figure(figsize=(16, 8), constrained_layout=False)
gs = fig.add_gridspec(nrows=2, ncols=2, height_ratios=[0.12, 1.0])


ax_legend = fig.add_subplot(gs[0, :])
ax_legend.axis('off')


legend_handles = [
    Patch(facecolor=my_cmap(rescale_for_cmap(e)), edgecolor='black', label=epsilon_label(i, e))
    for i, e in enumerate(epsilon_center)
]

legend = ax_legend.legend(
    handles=legend_handles, title=r'$\varepsilon$',
    loc='center', ncol=min(6, len(legend_handles)),
    frameon=True, fontsize=8, title_fontsize=9,
    borderpad=0.2, labelspacing=0.2, handlelength=0.8, handletextpad=0.3, columnspacing=0.8
)


ax_psha = fig.add_subplot(gs[1, 0], projection='3d')
ax_plha = fig.add_subplot(gs[1, 1], projection='3d')


bottom = np.zeros(len(x))
for i, e in enumerate(epsilon_center):
    top = disagg_psha[pga_index, :, :, i].ravel()
    ax_psha.bar3d(
        x, y, bottom, width, depth, top,
        shade=True, zsort='max', edgecolor='black',
        facecolor=my_cmap(rescale_for_cmap(e))
    )
    bottom += top

ax_psha.set_title(f'a) PSHA Disaggregation  PGA = {np.round(pga[pga_index], 3)} g', pad=6)
ax_psha.set_xlabel('Distance [km]')
ax_psha.set_ylabel('Magnitude')
ax_psha.set_zlabel('% Contribution')


bottom = np.zeros(len(x))
for i, e in enumerate(epsilon_center):
    top = disagg_plha[fsl_index, :, :, i].ravel()
    ax_plha.bar3d(
        x, y, bottom, width, depth, top,
        shade=True, zsort='max', edgecolor='black',
        facecolor=my_cmap(rescale_for_cmap(e))
    )
    bottom += top

ax_plha.set_title(
    f'b) PLHA Disaggregation  FSL = {np.round(fsl[fsl_index], 3)} (PGA = {np.round(pga[pga_index], 3)} g)',
    pad=6
)
ax_plha.set_xlabel('Distance [km]')
ax_plha.set_ylabel('Magnitude')
ax_plha.set_zlabel('% Contribution')



plt.tight_layout()

plt.show()
```
